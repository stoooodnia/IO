```python
import pandas as pd

miasta = pd.read_csv('./miasta.csv')
print(miasta)
```

```python
nowyWiersz = {'Rok': '2010','Gdansk':'460','Poznan':'555','Szczecin':'405'}
miasta = miasta.append(nowyWiersz, ignore_index=True)
print(miasta)
```

```python
import matplotlib.pyplot as plt
x = miasta['Rok'].values
y = miasta['Gdansk'].values
print(x)
print(y)
plt.plot(x,y,'r',marker='.') 
plt.title('Liczba ludnosci w Gdansku')
plt.xlabel('Rok')
plt.ylabel('Liczba ludnosci')
plt.show()
```

```python
xp = miasta['Rok'].values
yp = miasta['Poznan'].values

xs = miasta['Rok'].values
ys = miasta['Szczecin'].values

plt.plot(x,y,'r',marker='.')
plt.annotate('Gdansk', xy=(2000, 460), xytext=(2000, 480))
plt.plot(xp,yp,'b',marker='.')
plt.annotate('Poznan', xy=(2000, 555), xytext=(2000, 585))
plt.plot(xs,ys,'g',marker='.')
plt.annotate('Szczecin', xy=(2000, 405), xytext=(2000, 355))

plt.xlabel('Rok')
plt.ylabel('Liczba ludnosci')
plt.title('Liczba ludnosci w miastach')
plt.show()
```

```python
import sklearn.preprocessing as skp

scaler = skp.StandardScaler()

scaled_data = scaler.fit_transform(miasta)
miasta_standarized = pd.DataFrame(scaled_data, columns=miasta.columns)

print("\nUstandaryzowane dane:\n", miasta_standarized)
```

przeskalowane dane będą miały średnią równą 0 i odchylenie standardowe równe 1.

```python
nScaler = skp.MinMaxScaler()
nScaled_data = nScaler.fit_transform(miasta)
miasta_normalized = pd.DataFrame(nScaled_data, columns=miasta.columns)

print("\nZnormalizowane dane:\n", miasta_normalized)
```

Po takiej normalizacji dane mają wartości z zakresu 0-1
